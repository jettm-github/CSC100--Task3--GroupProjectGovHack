```python
import pandas as pd
```

```python
input_path: str = "./Alpha_X_Aggregated.csv"

# read csv
df = pd.read_csv(input_path)
```

```python
df.head()
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
# List of columns to drop
columns_to_drop = ['dev_id']

# Drop the specified columns from the DataFrame
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Summing columns by transportation type into the 'df' DataFrame
df['Total People'] = df['F1 - peopleLeft/Up'] + df['F2 - peopleRight/Down']
df['Total Bikes']  = df['F3 - bikesLeft/Up']  + df['F4 - bikesRight/Down']
df['Total Cars']   = df['F5 - carsLeft/Up']   + df['F6 - carsRight/Down']
df['Total Buses']  = df['F7 - busesLeft/Up']  + df['F8 - busesRight/Down']
df['Total Trucks'] = df['F9 - trucksLeft/Up'] + df['F10 - trucksRight/Down']
```

```python
df.head()
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
clean_df = df.dropna()
```

```python
rel_output_file_path: str = "../../api/db/data/sunshine_coast_council_alpha_x.csv"
clean_df.to_csv(rel_output_file_path, index=False)
```
